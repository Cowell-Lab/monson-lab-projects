```python
with open('cat3-7groups_for_paper.airr.json', 'r') as file:
    airr_md = json.load(file)
orig_groups = airr_md.get('RepertoireGroup')

for l in [14, 15, 18, 19, 22, 23]:
    temp_groups = copy.deepcopy(orig_groups)
    for g in temp_groups:
        g['repertoire_group_id'] = g.get('repertoire_group_id')+'-'+str(l)
        g['repertoire_group_name'] = g.get('repertoire_group_name')+' RF'+str(l)
        g.setdefault('filter', {})
        g['filter']['Rearrangement'] = {
            "op": "=",
            "content": {
                "field": "junction_aa_length",
                "value": l
            }
        }
        # print(l, g.get('repertoire_group_name'))

    airr_md['RepertoireGroup'] = airr_md.get('RepertoireGroup') + temp_groups
        
with open('cat3-7groups_with_RF.json', 'w') as f:
    f.write(json.dumps(airr_md, indent=2))
```
