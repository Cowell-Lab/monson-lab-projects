```python
def find_duplicates(reps):
    seen = set(reps['subjects'])
    duplicates = set()
    for item in reps:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return list(duplicates)

find_duplicates(data['Repertoire'])
```
